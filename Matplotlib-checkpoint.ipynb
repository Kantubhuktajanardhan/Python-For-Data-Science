```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(10)
```

```python
y1=x**2
y2=2*x+3
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
themes=plt.style.available
print(themes)
```

```python
plt.style.use('seaborn-v0_8')
```

```python
plt.plot(x, y1, color='red', marker='o', label="iPhone")
plt.plot(x, y2, color='green', marker="*", markersize =12, linestyle = "dashed", label="Android")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Smartphones over time")
plt.legend()
plt.show()
```
